```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3]
y = [1, 2, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("X and Y sample")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["this is y", "this is z"])
plt.show()
```
